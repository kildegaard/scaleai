First prompt:

I want two agents, one doctor, and another patient. Use two LLMs for two different agents and loop the two agents in such a way, that if a patient asks a question, the doctor should respond back to the patient and the loop continues till the patient is satisfied. Using langchain agents and Llama and Mistral models create a Python code.

Here is a Python code that uses two Langchain agents, one as a doctor and the other as a patient. The doctor agent uses the LLaMA model, and the patient agent uses the Mistral model. The loop continues until the patient is satisfied.

```python
import langchain
from langchain.agents import load_tools
from langchain.agents import initialize_agent
from langchain.llms import LLaMA, Mistral

# Load the models
llama = LLaMA(model_name="llama-7b", temperature=0)
mistral = Mistral(model_name="mistral-small")

# Create the doctor agent
doctor_tools = load_tools(["final_answer"])
doctor_agent = initialize_agent(tools=doctor_tools, llm=llama, agent="react-description", verbose=True)

# Create the patient agent
patient_tools = load_tools(["final_answer"])
patient_agent = initialize_agent(tools=patient_tools, llm=mistral, agent="react-description", verbose=True)

# Initialize the conversation
conversation = []

# Loop until the patient is satisfied
while True:
    # Get the patient's question
    patient_question = patient_agent.call(query="What is your question?")

    # Add the patient's question to the conversation
    conversation.append(patient_question)

    # Get the doctor's response
    doctor_response = doctor_agent.call(query=patient_question)

    # Add the doctor's response to the conversation
    conversation.append(doctor_response)

    # Check if the patient is satisfied
    patient_satisfied = patient_agent.call(query="Are you satisfied with the answer?")

    # If the patient is satisfied, break the loop
    if patient_satisfied.lower() == "yes":
        break

# Print the conversation
for i in range(0, len(conversation), 2):
    print("Patient:", conversation[i])
    print("Doctor:", conversation[i+1])
    print()
```
This code creates two agents, one for the doctor and one for the patient. The doctor agent uses the LLaMA model, and the patient agent uses the Mistral model. The patient agent asks a question, and the doctor agent responds. The loop continues until the patient is satisfied.

Note that this code assumes that the patient will respond with a simple "yes" or "no" to indicate whether they are satisfied with the answer. You may want to modify the code to handle more complex responses.

Also, keep in mind that this is just an example code and may not work perfectly in all cases. The performance of the models and the agents will depend on the specific use case and the quality of the models.

2nd prompt:

Now add one more agent /compounder who will collect money from the patient and ask questions first to the patient, then follow the question back to the doctor and the doctor again loops through the patient until it is satisfied. Update the code 

Here is the updated code with an additional agent for the compounder:

```python
import langchain
from langchain.agents import load_tools
from langchain.agents import initialize_agent
from langchain.llms import LLaMA, Mistral

# Load the models
llama = LLaMA(model_name="llama-7b", temperature=0)
mistral = Mistral(model_name="mistral-small")

# Create the doctor agent
doctor_tools = load_tools(["final_answer"])
doctor_agent = initialize_agent(tools=doctor_tools, llm=llama, agent="react-description", verbose=True)

# Create the patient agent
patient_tools = load_tools(["final_answer"])
patient_agent = initialize_agent(tools=patient_tools, llm=mistral, agent="react-description", verbose=True)

# Create the compounder agent
compounder_tools = load_tools(["final_answer"])
compounder_agent = initialize_agent(tools=compounder_tools, llm=mistral, agent="react-description", verbose=True)

# Initialize the conversation
conversation = []

# Loop until the patient is satisfied
while True:
    # Compounder asks questions to the patient
    compounder_question = compounder_agent.call(query="What brings you here today?")

    # Add the compounder's question to the conversation
    conversation.append(compounder_question)

    # Patient responds to the compounder's question
    patient_response = patient_agent.call(query=compounder_question)

    # Add the patient's response to the conversation
    conversation.append(patient_response)

    # Compounder asks for payment
    payment = compounder_agent.call(query="Please pay the consultation fee.")

    # Add the payment request to the conversation
    conversation.append(payment)

    # Patient pays the fee (just a placeholder, no actual payment is made)
    payment_response = patient_agent.call(query=payment)

    # Add the payment response to the conversation
    conversation.append(payment_response)

    # Compounder forwards the patient's response to the doctor
    doctor_query = compounder_agent.call(query="The patient says: " + patient_response)

    # Doctor responds to the patient's query
    doctor_response = doctor_agent.call(query=doctor_query)

    # Add the doctor's response to the conversation
    conversation.append(doctor_response)

    # Check if the patient is satisfied
    patient_satisfied = patient_agent.call(query="Are you satisfied with the answer?")

    # If the patient is satisfied, break the loop
    if patient_satisfied.lower() == "yes":
        break

    # If the patient is not satisfied, continue the loop
    else:
        # Compounder forwards the patient's dissatisfaction to the doctor
        doctor_query = compounder_agent.call(query="The patient is not satisfied with the answer.")

        # Doctor responds to the patient's dissatisfaction
        doctor_response = doctor_agent.call(query=doctor_query)

        # Add the doctor's response to the conversation
        conversation.append(doctor_response)

# Print the conversation
for i in range(0, len(conversation), 2):
    print("Compounder:", conversation[i])
    print("Patient/Doctor:", conversation[i+1])
    print()
```
This code adds a compounder agent that collects money from the patient and asks questions first to the patient. The compounder then forwards the patient's response to the doctor, and the doctor responds back to the patient. The loop continues until the patient is satisfied.

Note that this code assumes that the patient will respond with a simple "yes" or "no" to indicate whether they are satisfied with the answer. You may want to modify the code to handle more complex responses.

Also, keep in mind that this is just an example code and may not work perfectly in all cases. The performance of the models and the agents will depend on the specific use case and the quality of the models.
