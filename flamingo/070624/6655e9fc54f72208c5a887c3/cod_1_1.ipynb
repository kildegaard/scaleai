```python
import mysql.connector

db_name = "hospital"
username = "admin"
password = "your_password"

connection = mysql.connector.connect(
    host="localhost", database=db_name, user=username, password=password
)

# Create a cursor object
cursor = connection.cursor()


def create_tables():
    cursor.execute(
        """
    CREATE TABLE IF NOT EXISTS patients (
        patient_id INT AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(255) NOT NULL,
        age INT,
        phone_number VARCHAR(20)
    )
    """
    )

    cursor.execute(
        """
    CREATE TABLE IF NOT EXISTS doctors (
        doctor_id INT AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(255) NOT NULL,
        specialization VARCHAR(255)
    )
    """
    )
    cursor.execute(
        """
    CREATE TABLE IF NOT EXISTS nurses (
        nurse_id INT AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(255) NOT NULL
    )
    """
    )
    cursor.execute(
        """
    CREATE TABLE IF NOT EXISTS prescriptions (
        prescription_id INT AUTO_INCREMENT PRIMARY KEY,
        patient_id INT,
        medication VARCHAR(255) NOT NULL,
        dosage VARCHAR(255),
        instructions TEXT,
        FOREIGN KEY (patient_id) REFERENCES patients(patient_id)
    )
    """
    )
    cursor.execute(
        """
    CREATE TABLE IF NOT EXISTS assignments (
        assignment_id INT AUTO_INCREMENT PRIMARY KEY,
        patient_id INT,
        doctor_id INT,
        nurse_id INT,
        FOREIGN KEY (patient_id) REFERENCES patients(patient_id),
        FOREIGN KEY (doctor_id) REFERENCES doctors(doctor_id),
        FOREIGN KEY (nurse_id) REFERENCES nurses(nurse_id)
    )
    """
    )
    connection.commit()


def add_patient(name, age, phone_number):
    insert_patient = (
        """INSERT INTO patients (name, age, phone_number) VALUES (%s, %s, %s)"""
    )
    cursor.execute(insert_patient, (name, age, phone_number))
    connection.commit()
    return cursor.lastrowid


def add_doctor(name, specialization):
    insert_doctor = """INSERT INTO doctors (name, specialization) VALUES (%s, %s)"""
    cursor.execute(insert_doctor, (name, specialization))
    connection.commit()
    return cursor.lastrowid


def add_nurse(name):
    insert_nurse = """INSERT INTO nurses (name) VALUES (%s)"""
    cursor.execute(insert_nurse, (name,))
    connection.commit()
    return cursor.lastrowid


def add_prescription(patient_id, medication, dosage, instructions):
    insert_prescription = """
    INSERT INTO prescriptions (patient_id, medication, dosage, instructions)
    VALUES (%s, %s, %s, %s)
    """
    cursor.execute(insert_prescription, (patient_id, medication, dosage, instructions))
    connection.commit()


def assign_doctor_and_nurse(patient_id, doctor_id, nurse_id):
    insert_assignment = """INSERT INTO assignments (patient_id, doctor_id, nurse_id)
    VALUES (%s, %s, %s)"""
    cursor.execute(insert_assignment, (patient_id, doctor_id, nurse_id))
    connection.commit()


create_tables()
```
