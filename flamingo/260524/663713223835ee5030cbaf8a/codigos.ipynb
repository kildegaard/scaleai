```python
def max_rectangle(grid):
    m, n = len(grid), len(grid[0])
    left, right, height = [0] * n, [n] * n, [0] * n
    max_area = 0
    max_rect = None
    for i in range(m):
        cur_left, cur_right = 0, n
        for j in range(n):
            if grid[i][j] == "x":
                height[j] += 1
            else:
                height[j] = 0
        for j in range(n):
            if grid[i][j] != "x":
                cur_left = j + 1
            else:
                left[j] = max(left[j], cur_left)
        for j in range(n - 1, -1, -1):
            if grid[i][j] != "x":
                cur_right = j
            else:
                right[j] = min(right[j], cur_right)
        for j in range(n):
            if height[j] > 0:
                w = right[j] - left[j]
                area = height[j] * w
                if area > max_area:
                    max_area = area
                    max_rect = (i, j, height[j], w)
    return max_area, max_rect


def calculate_fence_cost(grid, rect):
    m, n = len(grid), len(grid[0])
    i, j, h, w = rect
    cost = 0
    top = "|" * w
    bottom = "|" * w
    for k in range(j, j + w):
        if i > 0 and grid[i - 1][k] == "s":
            top = top[: k - j] + "||" + top[k - j + 1 :]
        if i + h < m and grid[i + h][k] == "s":
            bottom = bottom[: k - j] + "||" + bottom[k - j + 1 :]
    cost += top.count("||") * 2 + top.count("|")
    cost += bottom.count("||") * 2 + bottom.count("|")
    left = "|" * h
    right = "|" * h
    for k in range(i, i + h):
        if j > 0 and grid[k][j - 1] == "s":
            left = left[: k - i] + "||" + left[k - i + 1 :]
        if j + w < n and grid[k][j + w] == "s":
            right = right[: k - i] + "||" + right[k - i + 1 :]
    cost += left.count("||") * 2 + left.count("|")
    cost += right.count("||") * 2 + right.count("|")
    for k in range(i, i + h):
        for col in range(j, j + w):
            if grid[k][col] == "n":
                cost += 1
    return cost


grid = [
    list(".....n."),
    list("s.x..n."),
    list("...s..."),
    list(".n....."),
    list(".....s."),
]

area, rect = max_rectangle(grid)
cost = calculate_fence_cost(grid, rect)

print(f"Area: {area} squares")
print(f"Fence Cost: {cost} coins")

i, j, h, w = rect
for k in range(len(grid)):
    row = ""
    if i <= k < i + h:
        row += "||"
    else:
        row += "  "
    for col in range(len(grid[0])):
        if i <= k < i + h and j <= col < j + w:
            row += "||" if grid[k][col] == "s" or grid[k][col] == "n" else " ."
        else:
            row += grid[k][col]
    if i <= k < i + h:
        row += "||"
    print(row)
```

```python
def max_rectangle(matrix):
    def possible_fence_cost(holes, width):
        return (
            sum(
                2 if holes[i] == "s" or holes[i + width] == "s" else 1
                for i in range(width)
            )
            * 2
            + sum(2 if holes[i] == "n" else 1 for i in range(0, len(holes), width)) * 2
        )

    def is_valid(holes, width, height):
        for i in range(len(holes)):
            if holes[i] == "n" and (
                i - width < 0
                or i + width >= len(holes)
                or i % width == 0
                or i % width == width - 1
            ):
                return False
        return True

    rows, cols = len(matrix), len(matrix[0])
    max_area = 0
    min_cost = float("inf")
    optimal_fence = []

    for row_start in range(rows):
        for col_start in range(cols):
            for row_end in range(row_start + 1, rows + 1):
                for col_end in range(col_start + 1, cols + 1):
                    holes = [
                        matrix[i][j]
                        for i in range(row_start, row_end)
                        for j in range(col_start, col_end)
                    ]
                    width, height = col_end - col_start, row_end - row_start
                    area = width * height
                    if area > max_area and is_valid(holes, width, height):
                        fence_cost = possible_fence_cost(holes, width)
                        if fence_cost < min_cost or area > max_area:
                            max_area = area
                            min_cost = fence_cost
                            optimal_fence = [
                                (
                                    "||"
                                    if matrix[i][j] == "s" or matrix[i + 1][j] == "s"
                                    else "|"
                                )
                                for i in range(row_start, row_end - 1)
                                for j in range(col_start, col_end)
                            ]
                            optimal_fence += [
                                "||" if matrix[i][j] == "n" else "|"
                                for i in range(row_start, row_end)
                                for j in range(col_start, col_end - 1)
                            ]
    return max_area, min_cost, optimal_fence


matrix = [
    list(".....n."),
    list("s.x..n."),
    list("...s..."),
    list(".n....."),
    list(".....s."),
]
area, cost, fence = max_rectangle(matrix)

print("Area:", area, "squares")
print("Fence Cost:", cost, "coins")
print("Fence Layout:")
for i in range(len(fence) // 2):
    print(" ".join([fence[2 * i], fence[2 * i + 1]]))
```
